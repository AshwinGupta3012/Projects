```python
mlist=[mlist/10000 for mlist in range (1,9999)] #creates a list of x from 0.01 to 0.99
```

```python
max=0
curr=0
```

```python
#Sort for max accuracy

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.0001,random_state=0)
reg=LinearRegression()                  #linear regression is implied on training set
reg.fit(x_train,y_train)
reg.score(x_test,y_test)                #checks accuracy 

curr=reg.score(x_test,y_test) 

for i in mlist:
    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=i,random_state=0)
    
    reg=LinearRegression()              #linear regression is implied on training set
    reg.fit(x_train,y_train)
    
    reg.score(x_test,y_test) 
    
    accuracy=reg.score(x_test,y_test)  
    
    if(accuracy>=curr):
        max=i
        curr=accuracy
        
max=max*100
curr=curr*100
print(f"The maximum accuracy of {curr:1.4}% is at {max}% of train data")
```
